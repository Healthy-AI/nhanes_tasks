## Hypertension (López-Martínez et al., 2020)
**Reference**: López-Martínez et al. "An artificial neural network approach for predicting 
hypertension using NHANES data." Scientific Reports, 2020.

**Label**: Self-reported hypertension on medication OR SBP ≥140 mmHg OR DBP ≥90 mmHg

**Inclusion criteria**: Adults ≥20 years

**Deviations from reference study**:
- Reference restricted to NHANES 2007-2016 (n=24,434); we include all available cycles
- 5-fold stratified CV instead of train/test split
- ANN replaced by benchmark model suite

```python
import pandas as pd
import numpy as np
import os, pickle

full = pd.read_pickle("/Users/alva/Documents/nhanes_tasks/data/NHANES/NHANES_master.pkl")
```

```python
# --- Hypertension task ---
# Reference: López-Martínez et al. (2020), Scientific Reports
# Population: Adults ≥20, NHANES 2007-2016
# Label: Self-reported hypertension on medication OR SBP ≥140 OR DBP ≥90
# Deviation: All available cycles included; 5-fold stratified CV instead of train/test split

subset_hyp = full[full['RIDAGEYR'] >= 20].copy()

data_hyp = subset_hyp[['RIAGENDR', 'RIDAGEYR', 'RIDRETH1', 'BMXBMI', 'LBXSCR',
                        'SMQ020', 'DIQ010', 'BPQ020', 'BPQ040A', 'BPXSY1', 'BPXDI1']].copy()
data_hyp['BPQ040A'] = data_hyp['BPQ040A'].fillna(0)
data_hyp = data_hyp.dropna()

y_hyp = (
    ((data_hyp['BPQ020'] == 1) & (data_hyp['BPQ040A'] == 1)) |
    (data_hyp['BPXSY1'] >= 140) |
    (data_hyp['BPXDI1'] >= 90)
).astype(int).values

features_hyp = ['RIAGENDR', 'RIDAGEYR', 'RIDRETH1', 'BMXBMI', 'LBXSCR', 'SMQ020', 'DIQ010']
X_hyp = data_hyp[features_hyp].values

print(f"Hypertension: n={len(y_hyp)}, prevalence={y_hyp.mean():.3f}, features={X_hyp.shape[1]}")
```

```python
os.makedirs('/Users/alva/Documents/nhanes_tasks/data/tasks/', exist_ok=True)

task = {
    'X': X_hyp,
    'y': y_hyp,
    'features': features_hyp,
    'name': 'hypertension'
}

with open('/Users/alva/Documents/nhanes_tasks/data/tasks/hypertension.pkl', 'wb') as f:
    pickle.dump(task, f)

print("Saved.")
```
