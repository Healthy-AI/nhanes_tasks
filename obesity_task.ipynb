## Obesity in diabetes (Wang et al., 2025)
**Reference**: Wang et al. "A machine learning model for predicting obesity risk in patients 
with diabetes mellitus: analysis of NHANES 2007-2018." Frontiers in Public Health, 2025.

**Label**: BMI ≥ 30

**Inclusion criteria**: Adults ≥18 years with diabetes (self-report, FPG ≥126, or HbA1c ≥6.5%)

**Deviations from reference study**:
- Reference restricted to NHANES 2007-2018 (n=3,794); we include all available cycles (n=11,135)
- Prevalence consistent with reference (51.9% vs 57.0%)
- 5-fold stratified CV instead of 70/30 split

```python
import pandas as pd
import numpy as np
import os, pickle

full = pd.read_pickle("/Users/alva/Documents/nhanes_tasks/data/NHANES/NHANES_master.pkl")
```

```python
# --- Obesity in diabetes task ---
# Reference: Wang et al. (2025), Frontiers in Public Health
# Population: Adults ≥18 with diabetes, NHANES 2007-2018
# Label: BMI ≥ 30
# Deviation: All available cycles included; 5-fold stratified CV

subset_ob = full[full['RIDAGEYR'] >= 18].copy()

subset_ob = subset_ob[
    (subset_ob['DIQ010'] == 1) |
    (subset_ob['LBXSGL'] >= 126) |
    (subset_ob['LBXGH'] >= 6.5)
]

subset_ob['obesity'] = (subset_ob['BMXBMI'] >= 30).astype(int)

features_ob = [
    'RIAGENDR', 'RIDAGEYR', 'RIDRETH1', 'DMDMARTL', 'INDFMPIR',
    'LBXSAL', 'LBXSAPSI', 'LBXSASSI', 'LBXSCR', 'LBXSTR',
    'LBXSUA', 'LBXHGB', 'BPQ020', 'MCQ160C', 'LBXGH',
    'SLD010H', 'DR1TTFAT'
]

data_ob = subset_ob[features_ob + ['obesity']].dropna(subset=['obesity'])
X_ob = data_ob[features_ob].values
y_ob = data_ob['obesity'].values

print(f"Obesity in diabetes: n={len(y_ob)}, prevalence={y_ob.mean():.3f}, features={X_ob.shape[1]}")
```

```python
os.makedirs('/Users/alva/Documents/nhanes_tasks/data/tasks/', exist_ok=True)

task = {
    'X': X_ob,
    'y': y_ob,
    'features': features_ob,
    'name': 'obesity_dm'
}

with open('/Users/alva/Documents/nhanes_tasks/data/tasks/obesity_dm.pkl', 'wb') as f:
    pickle.dump(task, f)

print("Saved.")
```
