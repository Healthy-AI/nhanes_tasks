```python
import pandas as pd
import numpy as np
import os, pickle

full = pd.read_pickle("/Users/alva/Documents/nhanes_tasks/data/NHANES/NHANES_master.pkl")
```

```python
subset_bmi = full[full['RIDAGEYR'] >= 18].copy()


subset_bmi['bmi_category'] = pd.cut(
    subset_bmi['BMXBMI'], 
    bins=[0, 18.5, 25, 30, np.inf], 
    labels=[0, 1, 2, 3]
)

features_bmi = [
    'RIAGENDR', 'RIDAGEYR', 'RIDRETH1', 'DMDMARTL', 'INDFMPIR',
    'LBXSAL', 'LBXSAPSI', 'LBXSASSI', 'LBXSCR', 'LBXSTR',
    'LBXSUA', 'LBXHGB', 'BPQ020', 'MCQ160C', 'LBXGH',
    'SLD010H', 'DR1TTFAT'
]

data_bmi = subset_bmi[features_bmi + ['bmi_category']].dropna(subset=['bmi_category'])
X_bmi = data_bmi[features_bmi].values
y_bmi = data_bmi['bmi_category'].astype(int).values
print(f"BMI categories: n={len(y_bmi)}, features={X_bmi.shape[1]}")
print(pd.Series(y_bmi).value_counts().sort_index())
```

```python
task = {
    'X': X_bmi,
    'y': y_bmi,
    'name': 'bmi_staging',
    'type': 'multiclass'
}

with open('/Users/alva/Documents/nhanes_tasks/data/tasks/bmi_staging.pkl', 'wb') as f:
    pickle.dump(task, f)

print("Saved.")
```
