## HbA1c Regression (Ezz et al., 2026)
**Reference**: Ezz M, et al. "Machine Learning Framework for HbA1c Prediction: Data Enrichment, 
Cost Optimization, and Interpretability Through Stratified Regression and Multi-Stage Feature 
Selection." Diagnostics, 2026;16(4):607. doi: 10.3390/diagnostics16040607

**Label**: Continuous HbA1c (LBXGH, %)

**Inclusion criteria**: Adults ≥18 years, NHANES 2007–2020, non-missing HbA1c

**Deviations from reference study**:
- Reference uses 40 features via ICS+RFECV pipeline; we use 14 pre-selected features
- Reference n=42,208 after exclusions; our n varies by available cycles
- 5-fold CV instead of stratified regression CV

```python
import pandas as pd
import numpy as np
import os, pickle

full = pd.read_pickle("/Users/alva/Documents/nhanes_tasks/data/NHANES/NHANES_master.pkl")
```

```python
# --- HbA1c regression task ---
subset_hba1c = full[
    (full['RIDAGEYR'] >= 18)
].copy()
features_hba1c = [
'RIDAGEYR', 'RIAGENDR', 'RIDRETH1', 'BMXBMI', 'BMXWAIST',
'BPXSY1', 'BPXDI1', 'LBXSCH', 'LBXSTR', 'LBDHDD',
'LBXSGL', 'SMQ020', 'DMDEDUC2', 'INDFMPIR'
]
data_hba1c = subset_hba1c[features_hba1c + ['LBXGH']].dropna()
X_hba1c = data_hba1c[features_hba1c].values
y_hba1c = data_hba1c['LBXGH'].values
print(f"HbA1c regression: n={len(y_hba1c)}, mean={y_hba1c.mean():.2f}, std={y_hba1c.std():.2f}")
print(f"  Age: median={np.median(data_hba1c['RIDAGEYR']):.1f}, IQR=[{np.percentile(data_hba1c['RIDAGEYR'], 25):.1f}, {np.percentile(data_hba1c['RIDAGEYR'], 75):.1f}]")
print(f"  Female: {(data_hba1c['RIAGENDR']==2).mean()*100:.1f}%")
print(f"  Features: {X_hba1c.shape[1]}")
```

```python
task_hba1c = {
    'X': X_hba1c,
    'y': y_hba1c,
    'name': 'hba1c',
    'type': 'regression'
}

with open('/Users/alva/Documents/nhanes_tasks/data/tasks/hba1c.pkl', 'wb') as f:
    pickle.dump(task_hba1c, f)

print("Saved.")
```
