```python
import pandas as pd
import numpy as np
import os, pickle

full = pd.read_pickle("/Users/alva/Documents/nhanes_tasks/data/NHANES/NHANES_master.pkl")

subset = full[full['RIDAGEYR'] >= 18].copy()
```

```python

data = subset.dropna(subset=['LBXGH', 'LBXSGL']).copy()

def classify_diabetes(row):
    hba1c = row['LBXGH']
    fpg = row['LBXSGL']
    if hba1c >= 6.5 or fpg >= 126:
        return 2  # Diabetes
    elif hba1c >= 5.7 or fpg >= 100:
        return 1  # Prediabetes
    else:
        return 0  # Normal

data['diabetes_status'] = data.apply(classify_diabetes, axis=1)
print(data['diabetes_status'].value_counts())
print(f"Prevalens prediabetes: {(data['diabetes_status']==1).mean():.3f}")
print(f"Prevalens diabetes: {(data['diabetes_status']==2).mean():.3f}")
```

```python
features = [
    'RIDAGEYR', 'RIAGENDR', 'RIDRETH1', 'BMXBMI', 'BMXWAIST',
    'BPXSY1', 'BPXDI1', 'LBXSTR', 'LBDHDD', 'LBXSCH',
    'SMQ020', 'DMDEDUC2', 'INDFMPIR'
]

data_clean = data.dropna(subset=features)
X = data_clean[features].values
y = data_clean['diabetes_status'].values

print(f"n={len(y)}, klasser={dict(zip(*np.unique(y, return_counts=True)))}")
```

```python
task = {
    'X': X,
    'y': y,
    'name': 'diabetes_staging',
    'type': 'multiclass'
}

with open('/Users/alva/Documents/nhanes_tasks/data/tasks/diabetes_staging.pkl', 'wb') as f:
    pickle.dump(task, f)

print("Saved.")
```
